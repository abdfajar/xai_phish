```python
import pandas as pd
from scipy.stats import ttest_rel, wilcoxon, shapiro
import os

# Load the dataset
file_path = "NoFS_result.xlsx"  # Replace with your Excel file path
sheet_name = 0  # Replace with the sheet name or index if needed
output_file = "fp_CB_analysis_results.txt"

# Load the Excel file
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Specify the two columns to compare
column_1 = "fp_cb"  # Replace with the first column name
column_2 = "fp_ebm"  # Replace with the second column name

# Check if the columns exist in the dataset
if column_1 not in df.columns or column_2 not in df.columns:
    raise ValueError(f"One or both columns ({column_1}, {column_2}) are not present in the dataset.")

# Extract the columns
data1 = df[column_1].dropna()  # Drop missing values
data2 = df[column_2].dropna()  # Drop missing values

# Ensure both columns have the same number of data points
if len(data1) != len(data2):
    raise ValueError("The two columns must have the same number of data points after dropping missing values.")

# Descriptive Statistics
desc_stats1 = data1.describe()
desc_stats2 = data2.describe()

# Normality Test
stat1, p1 = shapiro(data1)
stat2, p2 = shapiro(data2)

# Perform the appropriate statistical test
if p1 > 0.05 and p2 > 0.05:
    test_name = "Paired t-test"
    stat, p_value = ttest_rel(data1, data2)
else:
    test_name = "Wilcoxon signed-rank test"
    stat, p_value = wilcoxon(data1, data2)

# Determine significance
alpha = 0.05  # Significance level
significant = p_value < alpha
better_column = column_1 if data1.mean() > data2.mean() else column_2

# Prepare the interpretation
interpretation = f"""
Comparison of {column_1} and {column_2}

Descriptive Statistics:
{column_1}:
{desc_stats1.to_string()}

{column_2}:
{desc_stats2.to_string()}

Normality Test (Shapiro-Wilk):
{column_1} p-value: {p1}
{column_2} p-value: {p2}

Statistical Inference:
Test Used: {test_name}
Test Statistic: {stat}
P-Value: {p_value}
Significant Difference: {"Yes" if significant else "No"}

Interpretation:
{"There is a significant difference between the two columns." if significant else "No significant difference was found."}
The better column is: {better_column if significant else "N/A"}
"""

# Save the results to a text file
with open(output_file, "w") as file:
    file.write(interpretation)

print(f"Analysis complete. Results saved to {output_file}.")
```

# Merged All File

```python
import os

# Directory containing the text files
input_folder = "C:/Users/DELL/S3UI/analysis"  # Replace with the correct folder path
output_file = "merged_analysis.txt"  # Name of the final merged file

# Check if the input folder exists
if not os.path.exists(input_folder):
    raise FileNotFoundError(f"The specified folder does not exist: {input_folder}")

# Open the output file in write mode
with open(output_file, "w", encoding="utf-8") as outfile:
    # Loop through all files in the folder
    for filename in os.listdir(input_folder):
        # Check if the file is a text file
        if filename.endswith(".txt"):
            file_path = os.path.join(input_folder, filename)
            try:
                # Open each text file and append its content to the output file
                with open(file_path, "r", encoding="utf-8") as infile:
                    outfile.write(f"----- Start of {filename} -----\n")  # Optional header
                    outfile.write(infile.read())
                    outfile.write(f"\n----- End of {filename} -----\n\n")  # Optional footer
            except Exception as e:
                print(f"Error reading {filename}: {e}")

print(f"All text files have been merged into {output_file}.")
```
