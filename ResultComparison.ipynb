```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the CSV files
file1 = "modelRFFS_RFrfe.csv"  # Replace with the first file path
file2 = "modelRFNoFS_results.csv"  # Replace with the second file path

# Load data into DataFrames
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Ensure the two DataFrames align on "Dataset Name"
merged_df = pd.merge(df1, df2, on="Dataset Name", suffixes=('_file1', '_file2'))

# Columns to compare
columns_to_compare = [
    "Number of Rows",
    "Number of Columns",
    "Accuracy",
    "Precision",
    "Recall",
    "False Positive Rate",
    "ROC AUC"
]

# Value-by-value comparison
comparison_results = []
for col in columns_to_compare:
    merged_df[f"{col}_difference"] = merged_df[f"{col}_file1"] - merged_df[f"{col}_file2"]

    # Collect row-wise value comparison
    comparison_results.append({
        "Metric": col,
        "Mean Difference": merged_df[f"{col}_difference"].mean(),
        "Max Difference": merged_df[f"{col}_difference"].max(),
        "Min Difference": merged_df[f"{col}_difference"].min()
    })

# Convert to DataFrame for reporting
comparison_summary_df = pd.DataFrame(comparison_results)

# Statistical descriptive comparison
statistical_description = {}
for col in columns_to_compare:
    description_file1 = merged_df[f"{col}_file1"].describe()
    description_file2 = merged_df[f"{col}_file2"].describe()
    statistical_description[col] = {
        "File1 Description": description_file1,
        "File2 Description": description_file2
    }

# Statistical inference comparison (t-test)
inference_results = []
for col in columns_to_compare:
    t_stat, p_value = ttest_ind(merged_df[f"{col}_file1"], merged_df[f"{col}_file2"], equal_var=False)
    inference_results.append({
        "Metric": col,
        "T-Statistic": t_stat,
        "P-Value": p_value
    })

# Convert to DataFrame for inference results
inference_summary_df = pd.DataFrame(inference_results)

# Save results to CSV
comparison_summary_df.to_csv("comparison_summary.csv", index=False)
inference_summary_df.to_csv("inference_summary.csv", index=False)

# Display the results to the console
print("Comparison Summary:")
print(comparison_summary_df)

print("\nInference Summary:")
print(inference_summary_df)

print("Comparison and inference analysis completed. Results saved as CSV files.")

```

```python
import pandas as pd

# Load the CSV files
file1 = "modelRFFS_RFrfe.csv"  # Replace with the first file path
file2 = "modelRFNoFS_results.csv"  # Replace with the second file path

# Load data into DataFrames
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Rename performance metric columns for df1 and df2
performance_metrics = ["Accuracy", "Precision", "Recall", "False Positive Rate", "ROC AUC"]

df1_renamed = df1.rename(columns={col: f"F1_{col}" for col in performance_metrics})
df2_renamed = df2.rename(columns={col: f"F2_{col}" for col in performance_metrics})

# Ensure the two DataFrames align on "Dataset Name"
merged_df = pd.merge(df1_renamed, df2_renamed, on="Dataset Name")

# Save merged DataFrame to CSV
merged_df.to_csv("merged_performance_metrics.csv", index=False)

# Display the first few rows of the merged DataFrame in the console
print("Merged Performance Metrics:")
print(merged_df.head())

print("Merged performance metrics saved as 'merged_performance_metrics.csv'.")
```
